```python
import pandas as pd
reac = pd.read_csv(r"C:\Users\Vaishnavi\Desktop\p'covigilance project data\REAC25Q4.txt",sep="$", low_memory=False)
outc = pd.read_csv(r"C:\Users\Vaishnavi\Desktop\p'covigilance project data\OUTC25Q4.txt", sep="$", low_memory=False)
```

```python
outc['outc_cod'].value_counts()
```

```python
# Creating binary serious variable — 1 if outcome is clinically serious, 0 otherwise
outc['serious']= outc['outc_cod'].isin(['HO','DE','LT','DS','CA','RI']).astype(int) 
```

```python
outc[['outc_cod','serious']].head()
```

```python
# Keeping only columns needed for analysis
import pandas as pd
demo_small = demo[['primaryid','age','sex']]
drug_small = drug[['primaryid','drugname']].drop_duplicates()
reac_small = reac[['primaryid','pt']]
outc_small = outc[['primaryid','serious']]
```

```python
# Collapsing outc_small to one row per patient
# max() ensures patient is flagged serious=1 if any outcome was serious
outc_small=outc_small.groupby('primaryid')['serious'].max().reset_index()
```

```python
print(outc_small.shape)
```

```python
# Extract primaryid column
# Check which values are duplicated
# Count how many duplicates exist
print(outc_small['primaryid'].duplicated().sum())
```

```python
print(outc_small['serious'].value_counts())
```

```python
# FAERS data is not a random sample of all adverse events.It is heavily skewed towards serious cases — because serious cases are more likely to be reported.
```

```python
# Merging all 4 datasets on primaryid using left join
# Left join ensures we keep all patients from demo_small as our base
# Patients missing from other tables are kept with NaN values
# Step 1 — attach drug information to each patient
# drug_small has many rows per patient so df1 will expand in rows
import pandas as pd
df1=pd.merge(demo_small,drug_small, on='primaryid', how='left')
# Step 2 — attach reaction information
# reac_small also has many rows per patient so df2 expands further
df2=pd.merge(df1, reac_small,on='primaryid',how='left')
# Step 3 — attach serious outcome flag
# outc_small has one row per patient so shape stays same
df3=pd.merge(df2, outc_small, on='primaryid', how='left')
```

```python
# Verifying merge results — shape, sample rows, and missing values
print(df3.shape)
print (df3.head())
print(df3.isnull().sum())
```

```python
 #dropped rows where serious was NaN because assuming a missing outcome as non-serious would introduce bias into our statistical analysis. dropped rows where serious was NaN because assuming a missing outcome as non-serious would introduce bias into our statistical analysis.

df3=df3.dropna(subset=['serious'])
```

```python
print(df3.shape)
```

```python
#The Rule

#Only drop rows where your critical analysis variable is missing.
#Never use dropna() blindly.
#dropna() without subset drops all rows with any null value — this would remove valid drug-reaction records just because demographic data was missing, unnecessarily reducing our sample size and introducing selection bias.
```

```python
#For each patient — you need to:

#Get their list of drugs
#Generate all 2-drug combinations
#Store each combination with their primaryid
```

```python
from itertools import combinations
# step1:Group drugs by patient
drug_groups= drug_small.groupby('primaryid') ['drugname'].apply(list)
```

```python
#step2: generate combinations for each patient
combo_list=[]
```

```python
from itertools import combinations

combo_list = []

for primaryid, drugs in drug_groups.items():
    for combo in combinations(drugs, 2):
        combo_list.append({
            'primaryid': primaryid,
            'drug1': combo[0],
            'drug2': combo[1]
        })

combo_df = pd.DataFrame(combo_list)
```
